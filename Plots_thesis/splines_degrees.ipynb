# Plot splines of different degrees

```python
import matplotlib.pyplot as plt
import numpy as np


from scipy.interpolate import BSpline
```

```python
#QUARTIC
k4 = 4
t4 = [0, 0, 0, 0, 0, 1, 2, 3, 4, 5, 5, 5, 5, 5]
c4= [0, 0, 0, 0, 1, 0, 0, 0, 0]
spl4 = BSpline(t4, c4, k4)
spl4(2.5)
```

```python

#CUBIC
k = 3
t = [0, 0, 0, 0, 1, 2, 3, 4, 4, 4, 4]
c= [0, 0, 0, 1, 0, 0, 0]
spl = BSpline(t, c, k)
spl(2.5)

```

```python
#QUADRATIC
k2 = 2
t2 = [0, 0, 0, 1, 2, 3, 3, 3 ]
c2 = [0, 0, 1, 0, 0]
spl2 = BSpline(t2, c2, k2)
spl2(2.5)
```

```python
#ORDER 1
k3 = 1
t3 = [0, 0, 1,2,2 ]
c3 = [0, 1, 0]
spl3 = BSpline(t3, c3, k3)
spl3(2.5)
```

```python
plt.figure()
fig, ax = plt.subplots(figsize=(12,4))
xx = np.linspace(0, 4, 50)
ax.plot(xx, spl(xx), 'b-', lw=4, alpha=0.7, label='Cubic BSpline')
xx2 = np.linspace(0, 3, 50)
ax.plot(xx2, spl2(xx2), 'b-', lw=4, alpha=0.7, label='Quadratic BSpline', color= 'green')
xx3 = np.linspace(0, 2, 50)
ax.plot(xx3, spl3(xx3), 'b-', lw=4, alpha=0.7, label='Order 1 BSpline', color= 'orange')
xx4 = np.linspace(0, 5, 50)
ax.plot(xx4, spl4(xx4), 'b-', lw=4, alpha=0.7, label='Quartic BSpline', color= 'red')
ax.grid(True)
ax.legend(loc='best')
plt.show()
```
